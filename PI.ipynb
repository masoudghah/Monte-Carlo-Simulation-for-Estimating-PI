```python
from typing import Tuple
import numpy as np
import matplotlib.pyplot as plt

def generate_points(total_points: int) -> Tuple[np.ndarray, np.ndarray]:
    x = np.random.uniform(low=-1, high=1, size=total_points)
    y = np.random.uniform(low=-1, high=1, size=total_points)
    return x, y

def draw_points(
        x: np.ndarray, 
        y: np.ndarray, 
        is_inside_circle: np.ndarray, 
        *, 
        pi_approximation: float
) -> None:
    plt.figure(figsize=(6, 6))
    plt.scatter(x, y, c=is_inside_circle, cmap='coolwarm', alpha=0.5)
    plt.axis('equal')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Approximating PI with {len(x)} points: {pi_approximation: .4f}')
    plt.show()

total_points = 5
x, y = generate_points(total_points)
print(x)
print(y)
print((x ** 2) + (y ** 2))
print((x ** 2) + (y ** 2) <= 1)

```

```python
for total_points in range(1000, 100_000, 20_000):
    x, y = generate_points(total_points)
    is_inside_circle = (x ** 2) + (y ** 2) <= 1
    num_points_inside_circle = np.sum(is_inside_circle)
    pi_approximation = 4 * (num_points_inside_circle / total_points)
    draw_points(x, y, is_inside_circle, pi_approximation=pi_approximation)
```
